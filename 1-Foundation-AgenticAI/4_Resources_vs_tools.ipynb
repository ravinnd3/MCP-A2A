# Two ways to enhance LLM Capabilities in Agentic AI - Rosources & Tools

**Resources :**

    * We can improve the LLMs Expertise with Providing it Resources
    * Resources give the agent data or context to think better.
    * In simple words, this mean shoving data relevant to the Questions into the Prompt.


=> There is a technique like RAG to get really smart at picking the relevant data
* RAG - Retrieval Augmented Generation. Its a design pattern that improves an LLM's performance by giving to external information sources during
generation

**Tools :**

    * Tools are actions the LLM can perform via external systems.
    * It performs LLM to carry-out Task Query a DB. Message other LLMs etc

**Tool Calling - In Theory**

* **Tool calling** allows an LLM to interact with external software or systems - like a browser, database, calculator, or local code executor - by generating structured outputs (usually JSON or function calls) that trigger real-world operations
