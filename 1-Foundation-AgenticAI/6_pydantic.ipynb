# Pydantic : Data Validation and Settings Management Made Easy

Pydantic is a data validation and settings management library using Python type annotations. 

It enforces type constraints and validates data automatically.

Why use Pydantic ?

* Enforces Type Safety and data validation
* Parses and converts input types automatically
* Integrates well with FastAPI, SQLModel and other frameworks

```python
from pydantic import BaseModel


# Validate the user data 
class User(BaseModel):
    id : int
    name:str


usr = User(id=1,name="Ravi")
print(usr)
```

```python
# use default values

class User(BaseModel):
    id : int
    name:str 
    signup_ts:str | None = None
    salary:float | None = None


usr = User(id="1",name="Ravi",salary=500000)
print(usr)
```

# Senario : you are building a backend service that receives JSON payloads from a front-end forms:


You need to:

Validate the email format

Ensure age is a positive integer

Ensure password meets minimum length requirements

Convert "30" to integer

Parse "2025-01-01" into a datetime.date

```python

{
  "username": "johndoe",
  "password": "secret123",
  "email": "john@example.com",
  "age": "30",
  "join_date": "2025-01-01"
}
```

```python
from pydantic import BaseModel, EmailStr , field_validator,ValidationError
from datetime import date

class UserRegistration(BaseModel):
    username: str
    password: str
    email: EmailStr
    age: int
    join_date: date

    @field_validator('age',mode='after')
    @classmethod
    def validate_age(cls, value):
        if value < 0:
            raise ValueError('Age must be positive')
        return value


    @field_validator('password')
    @classmethod
    def validate_password(cls, value):
        if len(value) < 10:
            raise ValueError('Age must be positive')
        return value
```

```python
incoming_data = {
  "username": "johndoe",
  "password": "secret12397879879",
  "email": "john@example.com",
  "age": "30",
  "join_date": "2025-01-01"
}

try:
    user = UserRegistration(**incoming_data)
    print("User registration successful:", user)
    print("User registration successful:", user.model_dump_json())

except ValidationError as e:
    print("Validation error:", e)
```

# Nested Model

```python
class Address(BaseModel):
    street: str
    city: str
    zip_code: str

class UserRegistrationWithAddress(UserRegistration):
    address: Address

incoming_data = {
    "username": "johndoe",
    "password": "secret123534323",
    "email": "john@example.com",
    "age": "30",
    "join_date": "2025-01-01",
    "address": {
        "street": "123 Main St",
        "city": "New York",
        "zip_code": "10001"
    }
}

try:
    userWithAdd = UserRegistrationWithAddress(**incoming_data)
    #print(user)
    print("JSON output:", userWithAdd.model_dump_json())
except ValidationError as e:
    print("âŒ Validation Error:")
    print(e)
```
