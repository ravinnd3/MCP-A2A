What Agents are Actually Are?

1. AI Agents are programs where LLM outputs controls the workflow.


* An Agent is an entity that perceives its envirinment, makes decisions and takes actions to achive goals.
In AI, an agent is typically a software system that exhibits intelligent behaviour.


Characteristics of an Agent:
* Automony : Operates without direct intervention.
* Perception: Receives input from its environment (via sensors or APIs)
* Decision-Making: Chooses actions based on goals, data, and past outcomes.
* Action : Interact with the environments (via actuators or software interfaces)
* Goal-driven: Behavior is guided by objectives. 

Agentic AI :


Agentic AI refres to AI systems that just dont respond to commands, but initiate actions , set goals, make plans, and adjust behavior dynamically.


Agentic AI is dfifferent from Traditional workflow.
* Tradional workflow : Workflow are systems where LLMs and toolls are Orchestrated through pre-defined code paths.

* Agents: Agents are systems where LLMs directly own processes and tools uses, maintaion control over they accomplish tasks

```python
# Start building the workflow
from openai import OpenAI

```

```python


client = OpenAI(api_key=os.getenv('OPEN_API_KEY'))

print("Using API key:", os.getenv("OPEN_API_KEY")[:10] + "..." if os.getenv("OPEN_API_KEY") else "No key found!")

response = client.chat.completions.create(
    model="gpt-5-nano-2025-08-07",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user",   "content": "what is the temperature in Patna, Bihar?"},
    ],
    stream=False
)

raw_response = response.choices[0].message.content

from IPython.display import Markdown,display
# This package is used for displaying markdown content in Jupyter notebooks

display(Markdown(f"### Response:\n{raw_response}"))
```

```python
questions = "Please propose a challenging questions to assess someones IQ. Respond only with questions"
messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user",   "content": questions},
    ]
```

```python
response = client.chat.completions.create(
    model="gpt-5-nano-2025-08-07",
    messages=messages
)
```

```python
raw_response_of_questons = response.choices[0].message.content
display(Markdown(f"### Questions:\n{raw_response_of_questons}"))
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user",   "content": raw_response_of_questons}
    ]
```

```python
raw_response_of_questons_answers = client.chat.completions.create(
    model="gpt-5-nano-2025-08-07",
    messages=messages
)
```

```python
display(Markdown(f"### Answers:\n{raw_response_of_questons_answers.choices[0].message.content}"))
```
